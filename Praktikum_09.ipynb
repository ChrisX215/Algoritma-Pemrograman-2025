# Penugasan 04 - Modul buatan sendiri

```python
%%writefile konversi.py
def celcius_ke_fahrenheit(c):
    """Mengubah suhu dari Celcius ke Fahrenheit"""
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
    """Mengubah suhu dari Fahrenheit ke Celcius"""
    return (f - 32) * 5/9
```

```python
import konversi

print("=== Program Konversi Suhu ===")

c = 25
f = 77

print(f"{c}°C = {konversi.celcius_ke_fahrenheit(c):.2f}°F")
print(f"{f}°F = {konversi.fahrenheit_ke_celcius(f):.2f}°C")
```

# Penugasan 06 - Studi kasus mini

```python
with open("laporan.py", "w") as f:
    f.write("""
def tampilkan_skor(total_soal, benar):
    salah = total_soal - benar
    print("\\n=== Ringkasan Kuis ===")
    print(f"Total Soal : {total_soal}")
    print(f"Jawaban Benar : {benar}")
    print(f"Jawaban Salah : {salah}")
    print(f"Skor Akhir : {(benar / total_soal) * 100:.2f}%")
""")
```

```python
import random
import laporan

print("=== Kuis Penjumlahan Sederhana ===")

total_soal = 5
benar = 0

for i in range(1, total_soal + 1):
    a = random.randint(1, 10)
    b = random.randint(1, 10)
    jawaban_benar = a + b

    print(f"Soal {i}: {a} + {b} = ?")
    try:
        jawaban = int(input("Jawaban Anda: "))
        if jawaban == jawaban_benar:
            print("✅ Benar!\n")
            benar += 1
        else:
            print(f"❌ Salah! Jawaban yang benar: {jawaban_benar}\n")
    except ValueError:
        print("⚠️ Input tidak valid! Dihitung salah.\n")

laporan.tampilkan_skor(total_soal, benar)
```
