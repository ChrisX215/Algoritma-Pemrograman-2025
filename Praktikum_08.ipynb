# Praktikum 04 – Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
    return sum(args)

def biodata(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

# Praktikum 07 - Fungsi decorator

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan...")
    fungsi()
    print("Setelah fungsi dijalankan...")
  return wrapper
  print(wrapper)

@decorator_sapa
def sapa():
  print("Halo mahasiswa Polines")

sapa()
```

# Penugasan 04 - Fungsi dengan *args

```python
def jumlahkan_semua(*angka):
    return sum(angka)

print(jumlahkan_semua(2, 4, 6))
print(jumlahkan_semua(1, 3, 5, 7, 9))
print(jumlahkan_semua(10, 20, 30, 40, 50, 60, 70))
```

#Penugasan 08 - Fungsi decorator

```python
def logger(fungsi):
    def wrapper():
        print("Memulai fungsi...")
        fungsi()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo, mahasiswa Polines!")

halo()
```

#Penugasan 09 - Studi Kasus Nyata – Program Nilai

```python
def input_nilai():
    nilai = []
    jumlah = int(input("Masukkan jumlah mahasiswa: "))
    for i in range(jumlah):
        n = float(input(f"Masukkan nilai mahasiswa ke-{i+1}: "))
        nilai.append(n)
    return nilai

def hitung_rata(nilai):
    return sum(nilai) / len(nilai)

def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak Lulus"

nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)

print("=== Program Nilai Mahasiswa ===")

data_nilai = input_nilai()

rata2 = hitung_rata(data_nilai)
print(f"\nRata-rata nilai: {rata2:.2f}")

print("Status:", status_kelulusan(rata2))

tugas = float(input("\nMasukkan nilai tugas: "))
ujian = float(input("Masukkan nilai ujian: "))
akhir = nilai_akhir(tugas, ujian)
print(f"Nilai akhir (40% tugas + 60% ujian): {akhir:.2f}")
```
