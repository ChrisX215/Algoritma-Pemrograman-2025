# Praktikum 01 - Bubble Sort

```python
def bubble_sort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
                return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

# Praktikum 02 - Selection Sort

```python
def selection_sort(data):
    n = len(data)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if data[j] < data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return data


angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))

```

# Praktikum 03 – Insertion Sort

```python
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

# Praktikum 04 – Studi Kasus: Mengurutkan Nilai Mahasiswa

```python
def bubble_sort_nilai(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j][1] < data[j + 1][1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

# Praktikum 05 – Perbandingan Waktu Eksekusi Sorting

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j] :
                arr[j+1] = arr[j]
                j -= 1
        arr[j+1] = key
```

```python
import random
import time

data = [random.randint(1, 1000) for _ in range(1000)]

print(f"Menguji sorting dengan {len(data)} data acak...\n")

for func in [bubble_sort, selection_sort, insertion_sort]:

    salinan = data.copy()

    start = time.time()

    func(salinan)

    end = time.time()

    durasi = end - start
    print(f"{func.__name__:<15} membutuhkan waktu {durasi:.5f} detik")
```

# Praktikum 06 – Sorting Data Barang (Nama & Harga)

```python
def bubble_sort_key(data, key_index = 0, reverse = False):
  n = len(data)
  for i in range (n):
    for j in range (0, n - i - 1):
      if reverse:
        if data[j][key_index] < data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
  return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

# Penugasan 01 - Eksperimen Bubble Sort

```python
def bubble_sort_descending(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            # Ubah tanda '>' menjadi '<' untuk descending
            if data[j] < data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

# Pengujian
angka = [64, 34, 25, 12, 22, 11, 90]
print("Bubble Sort Descending:", bubble_sort_descending(angka.copy()))
```

# Penugasan 02 - Eksperimen Selection Sort

```python
def selection_sort_max_to_end(data):
    n = len(data)
    # Loop dari posisi terakhir mundur ke awal
    for i in range(n - 1, 0, -1):
        max_index = i
        # Cari elemen terbesar di sub-array 0 sampai i
        for j in range(i):
            if data[j] > data[max_index]:
                max_index = j
        # Tukar elemen terbesar ke posisi akhir (i)
        data[i], data[max_index] = data[max_index], data[i]
    return data

# Pengujian
angka = [64, 25, 12, 22, 11]
print("Selection Sort (Max ke Akhir):", selection_sort_max_to_end(angka.copy()))
```

# Penugasan 03 - Eksperimen Insertion Sort

```python
def insertion_sort_string(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        # Python otomatis membandingkan string secara alfabetis
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

# Pengujian
nama_mahasiswa = ["Zara", "Andi", "Budi", "Citra", "Doni"]
print("Insertion Sort (String):", insertion_sort_string(nama_mahasiswa.copy()))
```

# Penugasan 05 - Analisis Kompleksitas

```python
import random

# --- Definisi Fungsi dengan Counter ---
def count_bubble(arr):
    count = 0
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            count += 1  # Hitung perbandingan
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return count

def count_selection(arr):
    count = 0
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            count += 1 # Hitung perbandingan
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return count

def count_insertion(arr):
    count = 0
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        # Perbandingan terjadi di dalam kondisi while
        while j >= 0:
            count += 1 # Hitung perbandingan
            if arr[j] > key:
                arr[j + 1] = arr[j]
                j -= 1
            else:
                break
        arr[j + 1] = key
    return count

# --- Eksekusi Analisis ---
print(f"{'Ukuran Data':<15} | {'Bubble':<10} | {'Selection':<10} | {'Insertion':<10}")
print("-" * 55)

for size in [10, 50, 100]:
    # Generate data acak
    data_asli = [random.randint(1, 1000) for _ in range(size)]

    c_bubble = count_bubble(data_asli.copy())
    c_select = count_selection(data_asli.copy())
    c_insert = count_insertion(data_asli.copy())

    print(f"{size:<15} | {c_bubble:<10} | {c_select:<10} | {c_insert:<10}")
```

# Penugasan 06 - Proyek Mini

```python
def sort_barang(data, key_idx, is_desc):
    # Menggunakan Bubble Sort yang fleksibel
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            tukar = False
            if is_desc:
                if data[j][key_idx] < data[j + 1][key_idx]: tukar = True
            else:
                if data[j][key_idx] > data[j + 1][key_idx]: tukar = True

            if tukar:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

# --- Main Program Proyek Mini ---
list_barang = []

# Input Data
print("=== Input Data Barang (Ketik 'stop' di nama untuk selesai) ===")
while True:
    nama = input("Nama Barang: ")
    if nama.lower() == 'stop':
        break
    try:
        harga = int(input("Harga Barang: "))
        list_barang.append((nama, harga))
    except ValueError:
        print("Harga harus angka!")

if list_barang:
    print("\n=== Menu Sorting ===")
    print("1. Nama (A-Z)")
    print("2. Nama (Z-A)")
    print("3. Harga (Termurah-Termahal)")
    print("4. Harga (Termahal-Termurah)")

    pilihan = input("Pilih menu (1-4): ")

    if pilihan == '1':
        hasil = sort_barang(list_barang, 0, False) # 0=Nama, False=Asc
    elif pilihan == '2':
        hasil = sort_barang(list_barang, 0, True)  # 0=Nama, True=Desc
    elif pilihan == '3':
        hasil = sort_barang(list_barang, 1, False) # 1=Harga, False=Asc
    elif pilihan == '4':
        hasil = sort_barang(list_barang, 1, True)  # 1=Harga, True=Desc
    else:
        hasil = list_barang
        print("Pilihan tidak valid")

    print("\nHasil Sorting:")
    for b in hasil:
        print(f"{b[0]}: Rp {b[1]}")
else:
    print("Tidak ada data barang.")
```
